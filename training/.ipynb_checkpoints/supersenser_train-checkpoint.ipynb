```python
data = pd.read_csv("germa_super_v16.tsv", sep="\t")
counts = pd.DataFrame.from_dict(Counter(data.Klasse), orient="index")
```

```python
gps = data.groupby("Klasse")
ix = np.hstack([np.random.choice(v, 900, replace=False) for v in gps.groups.values()])
```

```python
data = data.loc[ix,:]
```

```python
l_dict = {}
i = 0
for c in set(data["Klasse"]):
    l_dict[c] = i
    i+=1
```

```python
import json
with open("germa_dict.json","w") as f:
    json.dump(l_dict,f)
```

```python
X = data["wort"]
Y = np.array([l_dict[x] for x in data["Klasse"]])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.1,stratify=Y)
```

```python
X_train_vec = np.stack([ft_model[str(x)] for x in X_train])
X_test_vec = np.stack([ft_model[str(x)] for x in X_test])
```

```python
X_all = np.concatenate([X_train_vec,X_test_vec])
```

```python
Y_semi = np.concatenate([y_train,np.array([-1]*len(X_test))])
```

```python
len(set(y_train))
```

```python
weight = 0.8
model = ivis.Ivis(n_epochs_without_progress=20, supervision_weight=weight,k=30,model="maaten",embedding_dims=2)
```

```python
model.fit(X_all,np.array(Y_semi))
```

```python
vecs = model.transform(X_test_vec)
```
